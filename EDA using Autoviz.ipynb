## EDA(Exploratory data analysis via 'autoviz' library)

The purpose of this notebook is to show the all possible plots (automatic) using autoviz library for the heart attack dataset 

This data is about heart attack possibility of a person. I took this dataset from kaggle: https://www.kaggle.com/nareshbhat/health-care-data-set-on-heart-attack-possibility?select=heart.csv

Features description

1. age
2. sex
3. chest pain type (4 values)
4. resting blood pressure
5. serum cholestoral in mg/dl
6. fasting blood sugar > 120 mg/dl
7. resting electrocardiographic results (values 0,1,2)
8. maximum heart rate achieved
9. exercise induced angina
10. oldpeak = ST depression induced by exercise relative to rest
11. the slope of the peak exercise ST segment
12. number of major vessels (0-3) colored by flourosopy
13. thal: 0 = normal; 1 = fixed defect; 2 = reversable defect
14. target: 0= less chance of heart attack 1= more chance of heart attack

```python
import pandas as pd
df = pd.read_csv('heart.csv')
df.head()
```

```python
#Importing the autoviz_class library 
from autoviz.AutoViz_Class import AutoViz_Class

AV = AutoViz_Class()
```

```python
#defining the required valiables 
filename = "heart.csv"      #dataset
sep = ","
dft = AV.AutoViz( filename,
    sep=',',
    depVar='',
    dfte=None,
    header=0,
    verbose=0,
    lowess=False,
    chart_format='svg',
    max_rows_analyzed=150000,
    max_cols_analyzed=30,)
```

As we can see this library provide all the information of the dataset automatically we don't need to write code for each and every exploration of features.
